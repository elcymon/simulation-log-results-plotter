#  Information Visualization
The purpose of this notebook is to visualize the information logged during simulation.

## Robot Visualization
This section is to view the information logged by a robot. I think I can include other methods to visualize information from the whole swarm. But for now, I will stick to visualizing a single robot

```python
import pandas as pd
# filename = gui_fname()

# results_folder = '2018-11-16-TwoClusters-cap_com-qsize1-120-noise0-25pct/'
# filename = '../results/' + results_folder +  '100-RepAtt-Noise0pct-qsize120-2018-11-16--15-31-03_m_4wrobot14.txt'
# filename = '../1-RepAtt-Noise0pct-qsize1-2018-11-22--11-51-36_m_4wrobot4.txt'
def get_robot_data(filename):

    robot_data = pd.read_csv(filename,
                             header=None, sep=':|,',engine='python',
                             names=['robot','time','x','y','theta','turnP',
                                    'seen_litter','rep_neigh','rep_sig',
                                   'call_neigh','call_sig','collected_lit',
                                   'linear_dist','rot_dist','state'])
    return robot_data
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
!ls assets
```

```python
def visualize_robot_data(filename,title):
    robot_data = get_robot_data(filename)
    fig = plt.figure()
    plt.plot(robot_data['time'],robot_data['turnP'])
    
    plt.ylabel('Turn Probability')
    plt.xlabel('Time in seconds')
    plt.title(title + ': Noise=100, QSize=80')
    plt.ylim([0, robot_data['turnP'].max() * 1.2])
    fig.savefig(title + '.pdf',bbox_inches='tight')

    # robot_data.loc[520:550,['time','turnP','rep_sig','call_sig']]
    return robot_data.loc[:,['time','turnP','rep_sig','call_sig']]
```

```python
#!cp '../results/2018-11-22-TwoClusters-0-100pctNoise-1-40qSize/748-RepAtt-Noise100pct-qsize80-2018-11-22--14-32-04_m_4wrobot33.txt' ./assets/
filename = './assets/748-RepAtt-Noise100pct-qsize80-2018-11-22--14-32-04_m_4wrobot33.txt'
robot_info = visualize_robot_data(filename,'Updated-RepAtt-Experiment')
```
