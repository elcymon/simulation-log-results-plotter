## Visualize Effect of turnP
This notebook is to visualize the magnitude of angle turned and the rate of turning for a robot performing random walk in an unbounded world without any obstacles.

```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
from get_robot_data import get_robot_data
import numpy as np
```

```python
# !cp '../results/2018-11-26-NoLitter-test1/1-RW-0p0025P-2018-11-27--09-32-34_m_4wrobot1.txt' ./assets/
filename = './assets/1-RW-0p0025P-2018-11-27--09-32-34_m_4wrobot1.txt'
header = ['robot','time','x','y','theta','turnP','drd_heading','rnd_heading','linear_dist','rot_dist','action']
robot_data = get_robot_data(filename,header)
```

```python
robot_data
```

```python
robot_data2 = pd.DataFrame(columns=header)
# robot_data2 = robot_data2.append(robot_data.loc[0,:])
```

```python
p_action = robot_data.loc[0,'action']
# print(p_action)

for i in range(len(robot_data)-1):
    c_action = robot_data.loc[i + 1,'action']
    if p_action == 'straight' and c_action == 'turning':
        robot_data2 = robot_data2.append(robot_data.loc[i + 1,:])
#         print('appended')
    p_action = c_action
```

```python
robot_data2
```

```python
robot_data.loc[28340:28360,:]
```

```python
t = robot_data2.loc[:,'time']#extract the time column
tDiff = t.diff() # compute difference between turn times
# tDiff = tDiff.iloc[1:]# drop the first 
# robot_data2 = robot_data2.iloc[1:]# drop first row
# print(len(tDiff.loc[(tDiff<=10) & (tDiff >= 0)]))

robot_data2['turn_time'] = tDiff # create new column to record times turned
robot_data2.loc[robot_data2.index[0],'turn_time'] = robot_data2.iloc[0]['time'] # first turn is same as experiment time
# robot_data2.iloc[0]['turn_time'] = firstTurn
# print('No of time steps = {}, No of turns = {}'.format(len(robot_data),len(robot_data2)))
# print(tDiff)
# plt.plot(list(range(len(tDiff))),tDiff)
```

```python
robot_data2
```

```python
freqList = []
for i,tSize in enumerate([10,20,30,40]):
#     print(tSize)
    if tSize <= 30:
        freqList.append(len(tDiff.loc[(robot_data2['turn_time'] <= tSize) & (robot_data2['turn_time'] > (tSize - 10))]))
    else:
        freqList.append(len(tDiff.loc[(robot_data2['turn_time'] >= 30)]))
    
# print(freqList)
TDists = pd.DataFrame({'Frequency':freqList},index=['0-10','10-20','20-30','>30'])
straightTimeSteps = robot_data.loc[robot_data['action'] == 'straight']
print('Distribution of time till next turn\n',TDists)
print('Total time steps',len(robot_data))
print('Total turns made',len(robot_data2))
print('total steps / turns made',len(robot_data2) / len(robot_data))
print()
print('total straight time steps',len(straightTimeSteps))
print('total straight time steps / turns made',len(robot_data2) / len(straightTimeSteps))
print('turn probability',robot_data.loc[0,'turnP'])
```

```python
figname = './note-outputs/turn-freq-vs-time-rw-0p0025'
ax = TDists.plot.bar(figsize=(5,5),legend=False,rot=0)
ax.set_ylim(0,TDists['Frequency'].max() * 1.2)
ax.set_xlabel('Time in seconds')
ax.set_ylabel('Turn Counts')
for p in ax.patches:
    ax.annotate(str(int(p.get_height())), (p.get_x() +p.get_width()/2, p.get_height() + 0.5))

fig = ax.get_figure()
fig.savefig(figname + '.pdf', bbox_inches='tight')
```

```python
#Attempt to compute the linear distance between multiple turns using linear_dist column
lin_dist = robot_data2.loc[:,'linear_dist']
lin_distDiff = lin_dist.diff()
robot_data2['dist_btw_turn'] = lin_distDiff
robot_data2.loc[robot_data2.index[0],'dist_btw_turn'] = robot_data2.iloc[0]['linear_dist']

minDist = 0
maxDist = int(np.ceil(robot_data2['dist_btw_turn'].max()))
distIndexes = list(map(int, np.ceil(np.linspace(minDist,maxDist,10))))#create ramges as integers
distindexes = ['{}-{}'.format(i,j) for i,j in zip(distIndexes[0:-1],distIndexes[1:])]
distFrequencies = [] #frequency of occurence of a particular range
for i, dSize in enumerate(distindexes):
    dSize = dSize.split('-')
    dSize = list(map(int, dSize))
    distFrequencies.append(len(robot_data2.loc[(robot_data2['dist_btw_turn'] <= dSize[1])
                                              & (robot_data2['dist_btw_turn'] >= dSize[0])]))
```

```python
# rndHeading_Freq
figname = './note-outputs/path-length-distribution'
pathLengths = pd.DataFrame({'Frequency':distFrequencies},index=distindexes)
ax = pathLengths.plot.bar(legend = False,rot=0)
ax.set_ylim(0,pathLengths['Frequency'].max() * 1.2)
ax.set_xlabel('Path length in metres')
ax.set_ylabel('Counts')

for p in ax.patches:
    ax.annotate(str(int(p.get_height())), (p.get_x() +p.get_width()/4, p.get_height() + 0.5))
fig = ax.get_figure()
fig.savefig(figname + '.pdf', bbox_inches = 'tight')
```

```python
rnd_heading = robot_data2['rnd_heading']
rndHeading_Freq = []
angleSteps = 60
indexes = ['{}-{}'.format(i,j) for i,j in zip(np.arange(0,361,angleSteps),np.arange(angleSteps,361+angleSteps,angleSteps))]
indexes = indexes[0:-1]#remove last value (which is out of range 0 - 360)

for i,hSize in enumerate(indexes):
    hSize = hSize.split('-')
    hSize = list(map(int,hSize))
#     print(tSize)
    rndHeading_Freq.append(len(robot_data2.loc[(robot_data2['rnd_heading'] <= hSize[1])
                                                & (robot_data2['rnd_heading'] >= hSize[0])]))

```

```python
xx = ['{}-{}'.format(i,j) for i,j in zip(np.arange(0,361,angleSteps),np.arange(angleSteps,361+angleSteps,angleSteps))]
xx = xx[0:-1]
x = xx[0].split('-')
x = map(int,x)
list(x)
```

```python
# rndHeading_Freq
figname = './note-outputs/turn-angle-distribution'
rndHeadings = pd.DataFrame({'Frequency':rndHeading_Freq},index=indexes)
ax = rndHeadings.plot.bar(legend = False,rot=0)
ax.set_ylim(0,rndHeadings['Frequency'].max() * 1.2)
ax.set_xlabel('Angle Turned in degrees')
ax.set_ylabel('Counts')

for p in ax.patches:
    ax.annotate(str(int(p.get_height())), (p.get_x() +p.get_width()/2, p.get_height() + 0.5))
fig = ax.get_figure()
fig.savefig(figname + '.pdf', bbox_inches = 'tight')
```
